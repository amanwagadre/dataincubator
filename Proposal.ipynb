```python
import numpy as np
import csv

path = 'C:/Users/Aman/cmdbash/DataIncubator/crime_marijuana.csv'
with open(path, 'r') as f:
		reader = csv.reader(f, delimiter=',')
		# get header from first row
		headers = next(reader)
		# get all the rows as a list
		data = list(reader)
		# transform data into numpy array
		data = np.array(data)
		
[a,b]=data.shape

from geopy.geocoders import Nominatim

list1=[]

for i in range(0,a):
	try:
		geolocator = Nominatim(user_agent="specify_your_app_name_here")
		streetname=data[i,4]
		location = geolocator.geocode("{} CO".format(streetname))
		# print(location.address)
		# print(location.address,location.latitude, location.longitude,streetname)
		print(location.latitude, location.longitude)
		list1.append((location.latitude,location.longitude))
	except:
		None
	if i==100:
		break

###########################################################
from gmplot import gmplot

# Place map
gmap = gmplot.GoogleMapPlotter(39.6964327,-104.8752993, 13)


top_attraction_lats,top_attraction_lons=zip(*list1)
gmap.scatter(top_attraction_lats, top_attraction_lons, '#000000', size=200, marker=False)

# # Marker
# hidden_gem_lat, hidden_gem_lon = 37.770776, -122.461689
# gmap.marker(hidden_gem_lat, hidden_gem_lon, 'cornflowerblue')

# Draw
gmap.draw("map_denver.html")
```
